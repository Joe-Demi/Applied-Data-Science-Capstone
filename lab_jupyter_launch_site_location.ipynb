```python
# Task 2: Mark the success/failed launches for each site on the map
import folium

# Initialize a map centered at a global view
launch_map = folium.Map(location=[20, 0], zoom_start=2)

# List of launch sites with their coordinates and launch outcomes
launch_sites = [
    {"name": "Kennedy Space Center", "lat": 28.5721, "lon": -80.6480, "success": 90, "failed": 10},
    {"name": "Baikonur Cosmodrome", "lat": 45.965, "lon": 63.305, "success": 150, "failed": 20},
    {"name": "Vandenberg Space Force Base", "lat": 34.6328, "lon": -120.6108, "success": 50, "failed": 5},
    {"name": "Guiana Space Centre", "lat": 5.236, "lon": -52.768, "success": 60, "failed": 2},
    {"name": "Jiuquan Satellite Launch Center", "lat": 40.9606, "lon": 100.2983, "success": 80, "failed": 8}
]

# Add markers for each launch site showing successes and failures
for site in launch_sites:
    # Success marker (green)
    folium.Marker(
        [site["lat"], site["lon"]],
        popup=f'{site["name"]} - Successes: {site["success"]}',
        icon=folium.Icon(color='green', icon='ok-sign')
    ).add_to(launch_map)
    
    # Failed marker (red)
    folium.Marker(
        [site["lat"], site["lon"]],
        popup=f'{site["name"]} - Failures: {site["failed"]}',
        icon=folium.Icon(color='red', icon='remove-sign')
    ).add_to(launch_map)

# Save the map to an HTML file
launch_map.save('launch_sites_success_failures_map.html')

# To display the map in a Jupyter notebook
launch_map

```

```python
# TASK 3: Calculate the distances between a launch site to its proximities


from geopy.distance import geodesic

# Define the launch site coordinates (Kennedy Space Center as an example)
launch_site = (28.5721, -80.6480)  # Kennedy Space Center coordinates

# Define a list of nearby locations (with their coordinates)
proximities = [
    {"name": "Orlando", "coordinates": (28.5383, -81.3792)},
    {"name": "Cape Canaveral", "coordinates": (28.3922, -80.6077)},
    {"name": "Miami", "coordinates": (25.7617, -80.1918)},
    {"name": "Tampa", "coordinates": (27.9506, -82.4572)}
]

# Calculate distances between the launch site and nearby locations
for proximity in proximities:
    distance = geodesic(launch_site, proximity["coordinates"]).km
    print(f'Distance from {launch_site} to {proximity["name"]}: {distance:.2f} km')
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from geopy.distance import geodesic
from folium.features import DivIcon

# Define the launch site coordinates (Kennedy Space Center)
launch_site = (28.5721, -80.6480)

# Define the closest coastline coordinates
coastline = (28.56367, -80.57163)

# Calculate the distance between the launch site and the coastline
distance = geodesic(launch_site, coastline).km

# Initialize a folium map centered at the launch site
launch_map = folium.Map(location=launch_site, zoom_start=12)

# Add the launch site marker
folium.Marker(
    location=launch_site,
    popup="Kennedy Space Center",
    icon=folium.Icon(color="blue", icon="rocket")
).add_to(launch_map)

# Add the coastline marker
folium.Marker(
    location=coastline,
    popup="Closest Coastline",
    icon=folium.Icon(color="green", icon="info-sign")
).add_to(launch_map)

# Add a marker showing the distance between the launch site and the coastline
distance_marker = folium.Marker(
    location=coastline,
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance),
    )
)

# Add the distance marker to the map
distance_marker.add_to(launch_map)

# Save the map to an HTML file
launch_map.save('launch_site_to_coastline_map.html')

# If in a Jupyter notebook, display the map inline
launch_map
```

```python
import folium
from geopy.distance import geodesic
from folium.features import DivIcon

# Define the launch site coordinates (Kennedy Space Center)
launch_site = (28.5721, -80.6480)

# Define the closest coastline coordinates
coastline = (28.56367, -80.57163)

# Calculate the distance between the launch site and the coastline
distance = geodesic(launch_site, coastline).km

# Initialize a folium map centered at the launch site
site_map = folium.Map(location=launch_site, zoom_start=12)

# Add the launch site marker
folium.Marker(
    location=launch_site,
    popup="Kennedy Space Center",
    icon=folium.Icon(color="blue", icon="rocket")
).add_to(site_map)

# Add the coastline marker
folium.Marker(
    location=coastline,
    popup="Closest Coastline",
    icon=folium.Icon(color="green", icon="info-sign")
).add_to(site_map)

# Add a marker showing the distance between the launch site and the coastline
distance_marker = folium.Marker(
    location=coastline,
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance),
    )
)

# Add the distance marker to the map
distance_marker.add_to(site_map)

# Create a PolyLine between the launch site and the coastline
coordinates = [launch_site, coastline]
lines = folium.PolyLine(locations=coordinates, weight=2, color='blue')

# Add the PolyLine to the map
site_map.add_child(lines)

# Save the map to an HTML file
site_map.save('launch_site_to_coastline_map_with_line.html')

# If in a Jupyter notebook, display the map inline
site_map
```

```python
import folium
from geopy.distance import geodesic
from folium.features import DivIcon

# Define the launch site coordinates (Kennedy Space Center)
launch_site = (28.5721, -80.6480)

# Define the closest city (as an example)
closest_city = (28.5383, -81.3792)  # Orlando coordinates

# Calculate the distance between the launch site and the closest city
distance_to_city = geodesic(launch_site, closest_city).km

# Initialize a folium map centered at the launch site
site_map = folium.Map(location=launch_site, zoom_start=12)

# Add the launch site marker
folium.Marker(
    location=launch_site,
    popup="Kennedy Space Center",
    icon=folium.Icon(color="blue", icon="rocket")
).add_to(site_map)

# Add the marker for the closest city with distance information
city_marker = folium.Marker(
    location=closest_city,
    popup=f"Closest City: Orlando\nDistance: {distance_to_city:.2f} KM",
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#e74c3c;"><b>{:10.2f} KM</b></div>'.format(distance_to_city),
    )
)

# Add the city marker to the map
city_marker.add_to(site_map)

# Create a PolyLine between the launch site and the city
coordinates = [launch_site, closest_city]
line = folium.PolyLine(locations=coordinates, weight=2, color='red')

# Add the PolyLine to the map
site_map.add_child(line)

# Save the map to an HTML file
site_map.save('launch_site_to_city_map_with_line.html')

# If in a Jupyter notebook, display the map inline
site_map
```
